```python
import pandas as pd
import numpy as np
```

```python
data_2012 = pd.read_csv('BPI_2012.csv')
data_2012.head(15)
```

```python
data_2012['next_case'] = data_2012['case'].shift(-1)
data_2012['next_event'] = data_2012['event'].shift(-1)
for row, value in data_2012.iterrows():
    if value['case'] != value['next_case']:
        data_2012['next_event'][row] = 'NaN'
```

```python
data_2012.head()
```

```python
events = data_2012['event'].unique()
```

```python
df_copy = data_2012
```

```python
for x in events:
    df_copy[x] = df_copy.apply(lambda _: 0, axis=1)
```

```python
# making a list with all the case id's
lst_caseID = []
for x in data_2012['case']:
    lst_caseID.append(x)
lst_case_ID = []
for t in lst_caseID:
    if t not in lst_case_ID:
        lst_case_ID.append(t)
        
lst_case_ID
```

```python
lst_case_ID1 = lst_case_ID[:2]
lst_case_ID1
```

```python
for r in lst_case_ID:
    new_list = []
    for i in data_2012[data_2012['case'] == r].T:
        new_list.append(data_2012['event'][i])
        
        for (index, event) in enumerate(data_2012[data_2012['case'] == r].T):
            # Event is de huidige row, dus pas event aan
            occurred_events = new_list[0:index]
            for evType in occurred_events:
                data_2012.iloc[event, data_2012.columns.get_loc(evType)] = 1
```
