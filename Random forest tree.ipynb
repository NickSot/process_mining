```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data_2012 = pd.read_csv('BPI_2012.csv')
data_2012.head(15)
```

```python
data_2012.info()
```

```python
data_2012.describe()
```

```python
data_2012['nextEvent'] = data_2012['event']
```

```python
# making a list with all the case id's
lst_caseID = []
for x in data_2012['case']:
    lst_caseID.append(x)
lst_case_ID = []
for t in lst_caseID:
    if t not in lst_case_ID:
        lst_case_ID.append(t)
        
lst_case_ID
```

```python
# using the process id's to seperate the processes and in the processes making an extra collumn with the next event.
# for the last event there is NaN, because there is no next event in that process.
for i in lst_case_ID:
    for r in data_2012[data_2012['case'] == i].T:
        if r + 1 in data_2012[data_2012['case'] == i].T:
            data_2012['nextEvent'][r] = data_2012['event'][r+1]
        else:
            data_2012['nextEvent'][r] = 'NaN'
```

```python
# save the data, so you don't have to run the for loops above every time.
data_2012.to_csv('data_2012_with_nextEvent.csv')
```

```python
data_2012 = pd.read_csv('data_2012_with_nextEvent.csv')
```

```python
# making dummy variables from the event, in this way the model can use all the different event types.
event = pd.get_dummies(data_2012['event'])
event.head()
```

```python
# put the dummy variables back in the dataframe, this time as columns.
df2 = pd.concat([data_2012, event], axis=1)
df2.drop(['event'], axis=1, inplace=True)
df2.head()
```

```python
# dropping the rows, the model don't need.
from sklearn.model_selection import train_test_split
X = df2.drop(columns=['startTime', 'completeTime', 'REG_DATE', 'nextEvent', 'W_Valideren aanvraag'])
y = df2['nextEvent']
```

```python
# scaling all features 
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_scaled = sc.fit_transform(X)
```

```python
# split the test and train data, without shuffle. this is because the processes should stay with each other
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y,
                                                    test_size=0.3, shuffle=False)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(random_state=42)
rfc.fit(X_train, y_train)
rfc_pred = rfc.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
import seaborn as sns
cm =confusion_matrix(y_test, rfc_pred)

# Normalise
cmn = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
fig, ax = plt.subplots(figsize=(20,15))
sns.heatmap(cmn, annot=True, fmt='.2f', cmap='Blues', ax=ax)
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show(block=False)
```

```python
classification_report(y_test, rfc_pred, output_dict = True)
```

```python
from sklearn.model_selection import GridSearchCV
rfc_param_grid = {'n_estimators': [10, 50, 100, 200, 300, 500, 800],
                  'max_depth': [4, 5, 6, 7, 8],
                  'criterion': ['entropy', 'gini']}
rfc_grid = GridSearchCV(estimator=rfc,
                        param_grid=rfc_param_grid,
                        refit=True,
                        verbose = 3)
# fitting our model with optimized parameters
rfc_grid.fit(X_train,y_train)
# finding our best estimators
optimizedRFC = rfc_grid.best_estimator_
optimizedRFC
```

```python
import joblib
```

```python
joblib.dump(optimizedRFC,'./optimizedRFC.joblib')
```

```python
optimizedRFC= joblib.load('./optimizedRFC.joblib')
```

```python
# re-running preds
rfc_grid_pred = optimizedRFC.predict(X_test)
# re-evaluating
cm =confusion_matrix(y_test, rfc_grid_pred)

# Normalise
cmn = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
fig, ax = plt.subplots(figsize=(20,15))
sns.heatmap(cmn, annot=True, fmt='.2f', cmap='Blues', ax=ax)
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show(block=False)
```

```python
classification_report(y_test, rfc_grid_pred, output_dict = True)
```

```python
df_grouped = data_2012.groupby('case')
df_grouped
```

```python
list_1st_events = {'case_id': [], 'event': [], 'event_time': []}

for case, grp in df_grouped:
    sorted_grp = grp.sort_values('startTime')
    
    list_1st_events['case_id'].append(sorted_grp.iloc[0]['case'])
    list_1st_events['event'].append(sorted_grp.iloc[0]['event'])
    list_1st_events['event_time'].append(sorted_grp.iloc[0]['startTime'])
```

```python
first_events_times = pd.DataFrame(data=list_1st_events)
first_events_times.sort_values('event_time', inplace=True)

first_events_times
```

```python
first_events_times['time_threshold'] = pd.to_datetime('2011/12/31')

df_training_testing = first_events_times[first_events_times['event_time'] <first_events_times['time_threshold']]
df_validation = first_events_times[first_events_times['event_time'] >= first_events_times['time_threshold']]

df_training_testing = data_2012[data_2012['case'].isin(df_training_testing['case_id'])]
df_validation = data_2012[data_2012['case'].isin(df_validation['case_id'])]
```

```python
from sklearn.model_selection import train_test_split

df_train, df_test = train_test_split(df_training_testing, test_size=0.3)
```
